```python
plt.figure(figsize=(8,5))
plt.hist(df['age'],bins=30,edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['monthly_income'],bins=30,edgecolor='black')
plt.title("Monthly Income Distribution")
plt.xlabel("Monthly Income")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
df['shopping_preference'].value_counts().plot(kind='bar')
plt.title("Shopping Preference Distribution")
plt.xlabel("Shopping Preference")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(6,4))
df['gender'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['age'],df['monthly_online_orders'],alpha=0.3)
plt.title("Age vs Monthly Online Orders")
plt.xlabel("Age")
plt.ylabel("Monthly Online Orders")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['monthly_income'],df['avg_online_spend'],alpha=0.3)
plt.title("Income vs Avg Online Spend")
plt.xlabel("Monthly Income")
plt.ylabel("Avg Online Spend")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['monthly_online_orders'],df['monthly_store_visits'], alpha=0.3)
plt.title("Online Orders vs Store Visits")
plt.xlabel("Monthly Online Orders")
plt.ylabel("Monthly Store Visits")
plt.show()
```

```python
avg_orders = df.groupby('shopping_preference')['monthly_online_orders'].mean()
plt.figure(figsize=(6,4))
avg_orders.plot(kind='bar')
plt.title("Avg Online Orders by Shopping Preference")
plt.ylabel("Average Online Orders")
plt.xticks(rotation=0)
plt.show()
```

```python
city_pref = pd.crosstab(df['city_tier'],df['shopping_preference'])
city_pref.plot(kind='bar',figsize=(8,5))
plt.title("City Tier vs Shopping Preference")
plt.xlabel("City Tier")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```
